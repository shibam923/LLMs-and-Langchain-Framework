Question 1: What are Large Language Models (LLMs) and how do they function?

Answer:  Large Language Models (LLMs) are advanced artificial intelligence systems designed to understand, process, and generate human language. As an AI myself, I am powered by this exact underlying technology.

At their core, LLMs are highly sophisticated prediction engines. They do not "understand" text the way humans do; instead, they use complex mathematics and pattern recognition to determine which word is most likely to come next in a sequence.

### How Large Language Models Function

The process of building and running an LLM involves several critical phases:

**1. Massive Data Ingestion (Pre-training)**
LLMs are initially trained on vast amounts of text data scraped from the internet, including books, articles, websites, and conversational logs. This massive dataset acts as the foundation, allowing the model to learn grammar, facts, reasoning abilities, and the general structure of human language.

**2. Tokenization**
Before the AI can process human text, the words must be translated into a format computers can understand. The text is broken down into smaller pieces called "tokens." A token can be an entire word, a syllable, or even a single letter, which is then assigned a numerical value.

**3. The Transformer Architecture**
Most modern LLMs rely on a specific type of neural network called a "Transformer." The key innovation of the Transformer is the "self-attention mechanism." This allows the model to look at all the words in a sentence simultaneously and mathematically weigh their importance and relationships to one another, giving the model a deep understanding of context.

> **The Core Mechanism:** When you give an LLM a prompt, it uses its neural network to calculate the probability of the next token based on the context of all previous tokens. It builds its response one token at a time, very rapidly.

**4. Fine-Tuning and Alignment**
After the initial pre-training, the base model is essentially a massive autocomplete machine. To make it a helpful and safe conversational assistant, it undergoes "fine-tuning." This often involves a process like Reinforcement Learning from Human Feedback (RLHF), where human testers rank the AI's responses to teach it how to follow instructions, be polite, and avoid generating harmful content.




Question 2: Discuss the impact of LLMs on traditional software development
approaches.

Answer: The integration of Large Language Models (LLMs) into software development is arguably the biggest paradigm shift in the industry since the transition to high-level programming languages. We are moving away from an era where developers spend the majority of their time physically typing out syntax, and entering an era of orchestration and architectural design.

Here is a breakdown of how LLMs are fundamentally altering traditional software development approaches:

### 1. Accelerated Code Generation and "Agentic" Workflows

Traditionally, writing boilerplate code, setting up basic infrastructure, and fleshing out repetitive logic took up a massive portion of a developer's day.

* **Copilots as Standard:** Tools driven by LLMs (like GitHub Copilot, Claude Code, and others) act as highly advanced autocomplete engines, predicting entire blocks of code or functions based on a simple comment.
* **The Rise of Autonomous Agents:** We are currently seeing a shift toward "agentic" workflows. Instead of just generating lines of code, developers are increasingly orchestrating multiple AI agents to research, write, and refine entire software systems autonomously.

### 2. Redefining the Developer's Role: From "Coder" to "Architect"

Because LLMs can handle the heavy lifting of syntax generation, the core identity of the software engineer is shifting.

* **System Design Over Syntax:** Junior developers are increasingly expected to act like junior architects. The value of a developer now lies less in memorizing specific programming language quirks and more in understanding business requirements, system design, and how to piece together complex, AI-driven workflows.
* **Prompt Engineering as a Core Skill:** Natural language is becoming a legitimate programming interface. Knowing how to precisely prompt an LLM to generate secure, optimized, and maintainable code is now a critical engineering competency.

### 3. Transforming Testing and Debugging

Historically, writing unit tests and hunting down obscure bugs was a notoriously tedious part of the Software Development Life Cycle (SDLC).

* **Automated Test Generation:** LLMs can instantly generate comprehensive suites of unit tests, anticipating edge cases that a human developer might overlook.
* **Conversational Debugging:** Instead of solely relying on stack traces and print statements, developers can paste error logs into an LLM and have a conversational back-and-forth to diagnose the root cause and generate a patch.

### 4. New Challenges: The "Resiliency Imperative"

While LLMs drastically increase velocity, they introduce entirely new risks that traditional approaches didn't have to account for:

* **Hallucinations and Subtle Bugs:** LLMs can generate code that looks perfectly valid but contains subtle logical errors or security vulnerabilities. Developers must shift to a mindset of rigorous code review and validation.
* **Exploding Technical Debt:** Because it is so easy to generate thousands of lines of code in seconds, teams risk bloating their codebases. Managing the maintainability and cleanliness of AI-generated code is a massive new challenge.
* **Security Concerns:** Relying on AI agents means organizations must implement zero-trust architectures and ensure that generated code doesn't inadvertently introduce malicious vulnerabilities or leak proprietary data.


Question 3: What are the key advantages and limitations of using LLMs in real-world
applications?

Answer:  In real-world applications, Large Language Models offer transformative benefits but also introduce significant operational and ethical hurdles. Striking the right balance is the current frontier for businesses and developers alike.

Here is a breakdown of the key advantages and limitations of deploying LLMs today:

### Key Advantages

* **Versatility and Multimodal Capabilities:** Unlike traditional AI models built for a single, narrow task (like predicting housing prices), a single LLM can translate languages, write marketing copy, summarize 100-page legal documents, and debug software code. Modern models can also process and generate audio, images, and video, acting as comprehensive digital assistants.
* **Accelerated Productivity and Efficiency:** LLMs excel at automating cognitive heavy lifting. Tasks that previously took humans hours—such as extracting specific insights from massive datasets, drafting routine emails, or writing boilerplate code—can now be executed in seconds, freeing up human workers for high-level strategic thinking.
* **Intuitive Accessibility:** LLMs have democratized computing by making natural human language the primary programming interface. Non-technical users can perform complex data analysis or build software workflows simply by giving the AI plain-English instructions.
* **Scalability Across Industries:** Because they are pre-trained on vast amounts of general knowledge, LLMs can be easily "fine-tuned" or customized (using techniques like Retrieval-Augmented Generation, or RAG) for highly specific enterprise use cases, from healthcare diagnostics to financial fraud detection.

### Key Limitations

* **Hallucinations and Reliability Issues:** LLMs are fundamentally probability engines, not databases of facts. They can confidently generate highly convincing but completely fabricated information. Because they lack true cognitive understanding, they cannot independently verify the accuracy of their outputs, requiring rigorous human oversight.
* **Data Privacy and Security Risks:** Using public or cloud-based LLMs often requires feeding them data. If employees input proprietary code, sensitive patient records, or confidential financial data into a prompt, that information can potentially be absorbed into the model's training data or leaked. Furthermore, LLMs are vulnerable to "prompt injection" attacks, where bad actors manipulate the model's instructions.
* **Inherent Bias and Ethical Concerns:** LLMs learn from the internet, which means they absorb the historical biases, prejudices, and toxicities present in human-generated data. If left unchecked, these models can perpetuate discrimination in real-world applications, such as biased hiring algorithms or skewed loan approvals.
* **Massive Resource and Compute Costs:** The infrastructure required to train and run inference on large models is staggering. It demands thousands of specialized GPUs, high energy consumption, and significant financial investment. This cost and environmental footprint remain a major barrier to entry for smaller organizations.

Question 5: Compare and contrast Langchain and LamaIndex. What unique problems
does each solve?

Answer: When building applications powered by Large Language Models, developers quickly realize that raw LLMs need a supporting infrastructure to be genuinely useful. This is where frameworks like **LangChain** and **LlamaIndex** come in.

While they are often mentioned in the same breath, they are not direct competitors. Instead, they are complementary tools designed to solve entirely different sets of problems in the AI development lifecycle.

Here is a breakdown of how they compare and the unique challenges each addresses.

### High-Level Comparison

| Feature | LangChain (The Conductor) | LlamaIndex (The Librarian) |
| --- | --- | --- |
| **Core Purpose** | Orchestration, logic flows, and agentic behavior. | Data ingestion, indexing, and retrieval. |
| **Primary Focus** | What the AI *does* (tool use, sequencing, memory). | What the AI *knows* (connecting LLMs to private data). |
| **Ideal Use Case** | Autonomous agents, multi-step workflows, chatbots. | Enterprise search, document Q&A, RAG systems. |
| **Design Approach** | Broad, modular, and highly versatile. | Deep, focused, and optimized for data efficiency. |

---

### LangChain: Solving for "Action and Orchestration"

LangChain is a general-purpose framework designed to turn LLMs from passive text generators into active problem solvers. If you want your AI to take actions, LangChain provides the "brain wiring."

**Unique Problems LangChain Solves:**

* **Multi-Step Reasoning (Chaining):** LLMs typically handle one prompt at a time. LangChain solves this by allowing developers to link models together. For example, you can build a chain where step one extracts data, step two translates it, and step three drafts an email—all triggered by a single user input.
* **Tool Usage and Agentic Workflows:** LangChain gives LLMs hands. It allows models to autonomously decide when to use external tools, such as browsing the web, querying a weather API, or running a Python script to calculate math, rather than just guessing the answer.
* **Complex Memory Management:** Out of the box, LLMs have no memory. LangChain provides sophisticated memory modules (like summarizing past conversations or storing rolling chat histories) so chatbots can maintain long, coherent interactions with users without losing context.

### LlamaIndex: Solving for "Context and Knowledge"

LlamaIndex (formerly GPT Index) is laser-focused on Retrieval-Augmented Generation (RAG). While LangChain can do basic retrieval, LlamaIndex is purpose-built to efficiently feed your private, custom data to an LLM.

**Unique Problems LlamaIndex Solves:**

* **The Context Window Limit:** You cannot paste a 10,000-page enterprise database into a single AI prompt. LlamaIndex solves this by deeply integrating with hundreds of data sources (PDFs, Notion, SQL databases, Google Drive), breaking that data down ("chunking"), and structuring it into highly optimized, searchable indexes.
* **Precision Retrieval at Scale:** When a user asks a question, LlamaIndex excels at lightning-fast semantic search. It acts like a highly advanced librarian, instantly finding the exact three paragraphs across millions of documents needed to answer the user's specific query, and feeding *only* that relevant data to the LLM.
* **Reducing Hallucinations:** By mathematically grounding the AI's responses in your specific, indexed data, LlamaIndex significantly reduces the risk of the model making up false information.

---

### The Best of Both Worlds

In modern, production-grade AI applications, developers rarely choose just one. They are frequently used together to build highly capable systems.

A common architecture involves using **LlamaIndex** to fetch the highly accurate, relevant facts from a massive corporate database, and then passing that data to **LangChain**, which uses its agentic logic to format the answer, trigger a notification, and update a CRM system.



Question 6:  Implement a basic Langchain pipeline using OpenAI’s LLM to answer
questions based on a user input prompt.

(Include your Python code and output in the code box below.)

Answer:  

```python
import os
from google.colab import userdata # 1. Import Colab's secret manager
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 2. Fetch the secret and set it as an environment variable
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

def run_colab_pipeline(user_question):
    # LangChain will now automatically detect the GOOGLE_API_KEY
    # from the environment variable we just set above!
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

    template = """
    You are a highly intelligent and concise expert.
    Please answer the following question clearly and directly:

    Question: {question}

    Answer:
    """
    prompt = PromptTemplate.from_template(template)
    output_parser = StrOutputParser()

    # 3. Build and invoke the chain
    chain = prompt | llm | output_parser

    print(f"User: {user_question}\n")
    return chain.invoke({"question": user_question})

# --- Execution ---
sample_question = "What is the primary difference between a compiler and an interpreter?"
ai_response = run_colab_pipeline(sample_question)
print(f"AI:\n{ai_response}")
```

Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs.

(Include your Python code and output in the code box below.)

Answer:

```python
import os
import requests
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool

# NEW: Import the modern agent builder from LangGraph
from langgraph.prebuilt import create_react_agent

# 1. Securely load your API key
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

# 2. Define the Tool (Unchanged)
@tool
def get_current_weather(latitude: float, longitude: float) -> str:
    """Fetch the current weather for a specific latitude and longitude using the Open-Meteo API."""
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        current = data.get("current_weather", {})
        temp = current.get("temperature")
        windspeed = current.get("windspeed")
        return f"The current temperature is {temp}°C with a wind speed of {windspeed} km/h."
    else:
        return "Sorry, I could not fetch the weather data at this time."

def run_weather_agent(user_query: str):
    # 3. Initialize the Model
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

    # 4. Create the Agent using LangGraph
    # We no longer need a PromptTemplate or a separate AgentExecutor!
    tools = [get_current_weather]
    agent_executor = create_react_agent(llm, tools)

    # 5. Invoke the Agent
    # LangGraph uses a standard "messages" list to pass data
    response = agent_executor.invoke({"messages": [("user", user_query)]})

    # The final AI response is always the very last message in the list
    return response["messages"][-1].content

# --- Execution ---
sample_query = "Can you check the live weather in Tokyo? (Lat: 35.6895, Lon: 139.6917) and tell me if I need a heavy coat?"

print(f"User: {sample_query}\n")
final_answer = run_weather_agent(sample_query)
print(f"\nFinal AI Answer:\n{final_answer}")
```

Question 8: Create a LamaIndex implementation that indexes a local text file and
retrieves answers from it.

(Include your Python code and output in the code box below.)

Answer:  

```python
import os
import nest_asyncio
from google.colab import userdata
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# 0. Apply the patch to allow async code to run in Colab!
nest_asyncio.apply()

# 1. Securely load your Gemini API key
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

def run_hybrid_rag():
    # 2. Configure LlamaIndex (Hybrid Approach)
    Settings.llm = GoogleGenAI(model="gemini-2.5-flash", temperature=0)
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

    # 3. Create the data directory and sample file
    data_dir = "my_local_data"
    os.makedirs(data_dir, exist_ok=True)

    file_path = os.path.join(data_dir, "mars_facts.txt")
    with open(file_path, "w") as f:
        f.write(
            "Mars is the fourth planet from the Sun and the second-smallest planet in the Solar System. "
            "It is often referred to as the 'Red Planet' because the iron oxide prevalent on its surface "
            "gives it a reddish appearance. Mars has a thin atmosphere, and its surface features canyons, "
            "volcanoes, dry lake beds, and craters. Olympus Mons, the largest volcano in the solar system, "
            "is located on Mars. The planet has two small moons, Phobos and Deimos."
        )
    print(f"Created local file at: {file_path}\n")

    # 4. Load the Data
    print("Reading files into LlamaIndex...")
    documents = SimpleDirectoryReader(data_dir).load_data()

    # 5. Build the Index locally
    print("Building the Vector Index locally...")
    index = VectorStoreIndex.from_documents(documents)

    # 6. Create the Query Engine
    query_engine = index.as_query_engine()

    # 7. Ask a Question!
    user_question = "Why is it called the Red Planet, and what are the names of its moons?"
    print(f"\nUser: {user_question}")

    response = query_engine.query(user_question)
    return response

# --- Execution ---
if __name__ == "__main__":
    ai_response = run_hybrid_rag()
    print(f"\nAI Answer:\n{ai_response}")
```

Question 9: Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.

(Include your Python code and output in the code box below.)

Answer:

```python
import os
import nest_asyncio
from google.colab import userdata

# LlamaIndex Imports (The Librarian)
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.google_genai import GoogleGenAI as LlamaGoogleGenAI
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# LangChain / LangGraph Imports (The Agent/Orchestrator)
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import Tool
from langgraph.prebuilt import create_react_agent

# 0. Apply the Colab async patch and set the API key
nest_asyncio.apply()
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

def run_integrated_chatbot(user_query: str):
    # ==========================================
    # PART 1: THE LLAMAINDEX LIBRARIAN
    # ==========================================
    print("1. Initializing LlamaIndex knowledge base...")
    Settings.llm = LlamaGoogleGenAI(model="gemini-2.5-flash", temperature=0)
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

    data_dir = "my_local_data"
    os.makedirs(data_dir, exist_ok=True)

    file_path = os.path.join(data_dir, "mars_facts.txt")
    with open(file_path, "w") as f:
        f.write(
            "Mars is the fourth planet from the Sun and the second-smallest planet in the Solar System. "
            "It is often referred to as the 'Red Planet' because the iron oxide prevalent on its surface "
            "gives it a reddish appearance. Mars has a thin atmosphere, and its surface features canyons, "
            "volcanoes, dry lake beds, and craters. Olympus Mons, the largest volcano in the solar system, "
            "is located on Mars. The planet has two small moons, Phobos and Deimos."
        )

    documents = SimpleDirectoryReader(data_dir).load_data()
    index = VectorStoreIndex.from_documents(documents)
    query_engine = index.as_query_engine()

    # ==========================================
    # PART 2: THE BRIDGE (Wrapping LlamaIndex as a Tool)
    # ==========================================
    def search_mars_documents(query: str) -> str:
        """Searches the internal document database for facts about Mars."""
        response = query_engine.query(query)
        return str(response)

    # We convert the LlamaIndex query engine into a LangChain Tool
    mars_search_tool = Tool(
        name="Mars_Document_Search",
        description="Always use this tool to answer questions about Mars. Do not guess.",
        func=search_mars_documents
    )

    # ==========================================
    # PART 3: THE LANGCHAIN AGENT
    # ==========================================
    print("2. Booting up the LangChain Agent...")
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

    # We give the agent our LlamaIndex tool
    tools = [mars_search_tool]
    agent = create_react_agent(llm, tools)

    # 4. Run the Agent
    print(f"\nUser: {user_query}")
    response = agent.invoke({"messages": [("user", user_query)]})

    return response["messages"][-1].content

# --- Execution ---
if __name__ == "__main__":
    # Let's ask a specific question that requires the document
    sample_query = "What is the name of the largest volcano in the solar system, and where is it located?"

    ai_response = run_integrated_chatbot(sample_query)
    print(f"\nFinal AI Answer:\n{ai_response}")
```

Question 10: A legal firm wants to use AI to summarize large volumes of legal
documents and retrieve relevant information quickly. Propose a solution using
Langchain and LamaIndex, and explain how it would work in practice.

(Include your Python code and output in the code box below.)

Answer:   

 For a legal firm, precision is paramount. The stakes are incredibly high, and traditional "chatbot" approaches are dangerous because of the risk of AI hallucinations. Lawyers do not just need a chatbot; they need a highly deterministic system that can reliably compress massive volumes of text (summarization) and extract specific, accurate clauses (retrieval) without making anything up.

By combining LangChain and LlamaIndex, we can build a Dual-Routing Legal Agent.

How It Works in Practice

1. The Ingestion Engine (LlamaIndex): Legal documents (NDAs, master service agreements, case law) are ingested by LlamaIndex. Instead of just creating one database, LlamaIndex creates two separate mathematical indexes for the same documents:

   - A Vector Store Index: Chops the document into small, overlapping chunks. This is optimized for finding specific needles in the haystack (e.g., "What is the liability cap in section 4?").

   - A Summary Index: Links all the document chunks sequentially. This is optimized for synthesizing the entire document at once (e.g., "Give me a 3-paragraph summary of this entire merger agreement").

2. The Orchestrator (LangChain): Lawyers interact with a LangChain Agent. When a lawyer types a prompt, the LangChain Agent acts as a paralegal. It reads the prompt, decides whether the lawyer is asking for a broad summary or a specific fact, and routes the request to the correct LlamaIndex database.


```python
import os
import nest_asyncio
from google.colab import userdata

# LlamaIndex (The Legal Knowledge Base)
from llama_index.core import VectorStoreIndex, SummaryIndex, SimpleDirectoryReader, Settings
from llama_index.llms.google_genai import GoogleGenAI as LlamaGoogleGenAI
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# LangChain (The Paralegal Agent)
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import Tool
from langgraph.prebuilt import create_react_agent

# 0. Setup and Auth
nest_asyncio.apply()
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

def run_legal_agent(user_query: str):
    # ==========================================
    # 1. SETUP LLAMAINDEX & MOCK LEGAL DATA
    # ==========================================
    Settings.llm = LlamaGoogleGenAI(model="gemini-2.5-flash", temperature=0)
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

    data_dir = "legal_docs"
    os.makedirs(data_dir, exist_ok=True)

    # Let's create a mock Non-Disclosure Agreement (NDA)
    file_path = os.path.join(data_dir, "Acme_Corp_NDA_2024.txt")
    with open(file_path, "w") as f:
        f.write(
            "MUTUAL NON-DISCLOSURE AGREEMENT\n"
            "This Agreement is entered into between Acme Corp and Beta LLC.\n"
            "1. Confidential Information: Includes all technical and business data.\n"
            "2. Obligations: Receiving party must hold information in strict confidence for a period of 5 years.\n"
            "3. Exclusions: Information publicly known or independently developed is excluded.\n"
            "4. Governing Law: This agreement shall be governed by the laws of the State of Delaware.\n"
            "5. Penalty: Breach of this agreement carries a liquidated damages penalty of $500,000."
        )

    print("Ingesting legal documents...")
    documents = SimpleDirectoryReader(data_dir).load_data()

    # ==========================================
    # 2. BUILD THE DUAL INDEXES (LlamaIndex)
    # ==========================================
    print("Building Vector Index (for specific Q&A)...")
    vector_index = VectorStoreIndex.from_documents(documents)
    qa_engine = vector_index.as_query_engine()

    print("Building Summary Index (for document summarization)...")
    summary_index = SummaryIndex.from_documents(documents)
    # "tree_summarize" forces the AI to read the whole document, not just a snippet
    summary_engine = summary_index.as_query_engine(response_mode="tree_summarize")

    # ==========================================
    # 3. CREATE LANGCHAIN TOOLS
    # ==========================================
    def retrieve_specific_clause(query: str) -> str:
        """Retrieves specific clauses, facts, or data points from the legal documents."""
        return str(qa_engine.query(query))

    def summarize_document(query: str) -> str:
        """Generates comprehensive summaries of entire legal documents."""
        return str(summary_engine.query(query))

    # We give the agent explicit instructions on WHEN to use which tool
    tool_qa = Tool(
        name="Specific_Clause_Search",
        description="Use this tool ONLY when asked about specific clauses, governing laws, penalties, or targeted facts.",
        func=retrieve_specific_clause
    )

    tool_summary = Tool(
        name="Document_Summarizer",
        description="Use this tool ONLY when asked to summarize, outline, or provide an overview of a document.",
        func=summarize_document
    )

    # ==========================================
    # 4. RUN THE LANGCHAIN AGENT
    # ==========================================
    print("Booting up Legal Orchestrator Agent...\n")
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
    agent = create_react_agent(llm, [tool_qa, tool_summary])

    print(f"Lawyer: {user_query}\n")
    response = agent.invoke({"messages": [("user", user_query)]})

    return response["messages"][-1].content

# --- Execution ---
if __name__ == "__main__":
    # We will test the agent's ability to use BOTH tools for a complex request
    complex_query = "Please provide a brief summary of the Acme Corp NDA, and then explicitly tell me what the governing law and penalty are."

    ai_response = run_legal_agent(complex_query)
    print(f"AI Paralegal:\n{ai_response}")
```
